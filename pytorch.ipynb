```python
x_one=torch.ones_like(x_data)
print(x_one)
x_random=torch.rand_like(x_data,dtype=torch.float)
print(x_data)
print(x_random)
```

```python
shape=(2,3)
shape2=(3,4)
rand_tensor=torch.rand(shape)
rand1_tensor=torch.rand(shape2)
print(rand_tensor)
print(rand1_tensor)
```

```python
tensor=torch.rand(2,2)
t1=torch.cat((tensor,tensor))
print(tensor)
print(t1)
```

```python
tensor=torch.rand(4,4)
print(tensor)
print(tensor[:,2])
print(tensor[2])
print(tensor[2,2])
```

```python
tensor1=torch.rand(2,2)
tensor2=torch.rand(2,2)
y3=torch.matmul(tensor1,tensor2)
y4=tensor1.matmul(tensor2)
y5=tensor1@tensor2
print(y3)
print(y4)
print(y5)
#all these codes produce the exact answer
```

```python
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
y4=tensor.T

#tensor.T produces the transpose of the matrix



torch.matmul(tensor, tensor.T, out=y3)
print(f"y1 is:{y1}")
print(f"y2 is :{y2}")
print(f"y3 is :{tensor}")
print(f"y4 is :{y4}")
```

```python
agg=tensor.sum()
print(agg)
agg_value=round(agg.item(),3)
print(agg_value)
#here the .sum() function returns the sum of all elements in the tensor and the out is also a tensor
#if we need to strip down the tensor characters we need to use .item() which returns the normal value
```

```python
tensor=torch.ones(4,4)
print(f"{tensor}\n")
tensor.add_(5)#it adds in place that is no new tensor is created
tensor2=tensor.add(5)#a new tensor is created
print(tensor2)
print(tensor)
```

```python
#we can comfortably interchange between tensor and numpy
tensor=torch.ones(3,3)
print(tensor)
NpAcc=tensor.numpy()
print(NpAcc)
tensor2=torch.from_numpy(NpAcc)
print(tensor2)
```
